```python
# Creating age groups based on conditions
under_20 = epl_22[epl_22["Age"] < 20.0]
age_20_25 = epl_22[(epl_22["Age"] > 20.0) & (epl_22["Age"] <= 25.0)]
age_26_30 = epl_22[(epl_22["Age"] > 25.0) & (epl_22["Age"] <= 30.0)]
above_30 = epl_22[epl_22["Age"] > 30.0]
```

```python
x= np.array([under_20["Player"].count(), age_20_25["Player"].count(), age_26_30["Player"].count(), above_30["Player"].count()])
agelabels= ["<20","20-25","26-30","<30"]
plt.title('Player Age Groups', fontsize=14)
plt.pie(x, labels=agelabels, autopct='%0.01f%%')
plt.show()
```

```python
Young players (25 and under) make up 58.6% of the player base. 
This % is expected to keep growing in the coming years as academies are bringing in more talent and clubs are signing high potential youngsters
```

```python
# Average Age

plt.figure(figsize=(8, 6))
sns.boxplot(x="Team", y="Age",hue = "Team" ,data=epl_22, palette = "Set3")
plt.xticks(rotation=90)
plt.title("Distribution of Player Ages by Team")
plt.xlabel("Team")
plt.ylabel("Age")
plt.show()
```

```python
player_num = epl_22.groupby("Team").size()

data = epl_22.groupby("Team")["Age"].sum() / player_num
data.sort_values(ascending = False)
```

```python
Burnley has the highest average age (27.03) out of all the teams in the EPL. 
Arsenal have the lowest (22.48) which makes sense since they have been heavily investing in their youth players 
and have the highest number of under 20 players in their squad.
```

```python
# Sorting the Series by total assists per team
team_epl_22 = epl_22.groupby("Team")["Ast"].sum()
sorted_teams_by_assists = team_epl_22.sort_values(ascending=True)

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=sorted_teams_by_assists.index,hue = sorted_teams_by_assists.index, y=sorted_teams_by_assists.values, palette="Set2")
plt.xlabel("Team")
plt.ylabel("Total Assists")
plt.title("Total Assists by Each Team")
plt.xticks(rotation=90)
plt.show()
```

```python
sorted_by_goals = epl_22.groupby("Team")["Gls"].sum().reset_index()
top_10_teams = sorted_by_goals[["Team", "Gls"]].sort_values(by='Gls', ascending=True).head(10)
plt.figure(figsize=(10, 10))
sns.barplot(data=top_10_teams, x="Team", y="Gls",hue = "Team" , palette="rocket")
plt.xlabel("Team")
plt.ylabel("Goals")
plt.title("Top 10 Goals by Team")
plt.xticks(rotation=90)
plt.show()
```

```python
#Top 10 Goals produce by Player

top_10_player = epl_22[["Player","Team","Gls","MP"]].nlargest(n=10, columns='Gls')
top_10_player
```

```python
# Sort the DataFrame by 'Gls' (goals) in descending order and take the top 10 players
top_10_players = epl_22.sort_values(by='Gls', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_players, x="Player", y="Gls", hue = "Player", palette="Set3")
plt.xlabel("Players")
plt.ylabel("Total Goals")
plt.title("Top 10 Goal Scorers in EPL 2022-2023 Season")
plt.xticks(rotation=45, ha='right')  # Rotate x-labels for better readability
plt.tight_layout()  # Adjust layout for better appearance
plt.show()
```
